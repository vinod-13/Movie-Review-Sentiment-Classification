```python
#Code to Build Docker Image

import docker
client = docker.from_env()

result = client.images.build(
    path = 'C:\\Docker\\Demo',
    dockerfile = 'C:\\Docker\\Demo\\DockerFile',
    tag='python:0.1',
)

for item in result[1]:
    #print(item)
    for key, value in item.items():
        #print(key, ':', value)
        if key == 'stream':
            text = value.strip()
            if text:
                print(text)
```

```python
#Save to File as .tar
image = client.images.get("python:0.1")
f = open(r'C:\Docker\py.tar', 'wb')
for chunk in image.save():
    print("Image Save in Progress")
    f.write(chunk)
f.close()
print("Save Completed")
```

```python
#Code to Push the Image

client_to_push = docker.from_env()
client_to_push.login(username='******', password='******')
for line in client_to_push.images.push('ubuntu:latest', stream=True, decode=True):
    print(line)
```
